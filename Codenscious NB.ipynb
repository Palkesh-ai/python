# word frequency using strings

```python
#word frequency using strings
text = input("Enter a string: ")
words = text.split()
word_freq = {}
for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1
print("Word Frequency:", word_freq)
```

# Arrays and list

Check duplicates in list

```python
def has_duplicates(ls):
    if len(ls) != len(set(ls)): 
        return 1
    return 0 

my_ls = [1,2,3,4,4,5]
print(has_duplicates(my_ls))
```

Rotate by k

```python
def rotate(arr, k):
    n = len(arr)
    k= k%n
    return arr[k:] + arr[:k]
my_array = [1,2,3,4,5]
k = 1
print(rotate(my_array, k))
```

Frequency counter using list.

```python
def freq(arr):
    dictionary ={}
    for i in arr:
        if i in dictionary:
            dictionary[i] += 1
        else:
            dictionary[i] = 1
    return dictionary
my_arr = [1,2,2,3,4,4,5,5,5]
print(freq(my_arr))
```

# Strings

Check palindrom string

```python
def palindrome(s):
    reverse = s[::-1]
    if s == reverse:
        print("Palindrome")
    else:
        print("Not a palindrome")

str = input("Enter a string: ")
palindrome(str)
```

Word frequency counter

```python
def fr(s):
    word = s.lower().split()
    freq_dict = {}

    for w in word:
        if w in freq_dict:
            freq_dict[w] += 1
        else:
            freq_dict[w] = 1
    return freq_dict

s = input("Enter a string: ")
print(fr(s))

```

# Recursion

Sum of digits

```python
n = int(input("Enter a number: "))
def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return n%10 + sum_of_digits(n//10)
    
print("Sum of digits:", sum_of_digits(n))
```

reverse a string

```python
def rev(s):
    if len(s) == 0:
        return s
    else:
        return rev(s[1:]) + s[0]
string = input("Enter a string: ")
print("Reversed string:", rev(string))
```
