Conditionals

```python
age = int(input("Enter your age: "))
if age >= 18:
    print('you can vote')
elif age == 17:
    print('you can learn to drive')
else:
    print('Outside the if statement')
```

Loops in py

```python
for i in range(5,11,3):
    print(i)

a = ['one', 'two', 'three']

for ch in a:
    print(ch)
```

```python
#while loop
count = False
x = 0

while not count:
    if x == 5:
        count = True
    print ('print x')
    x += 1
```

we can use 'break' if we want to force stop loop.
