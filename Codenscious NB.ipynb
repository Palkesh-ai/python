# palindrom

```python
#code for palindrome using string

str = input("Enter a string: ")

rev_str = str[::-1]
#print("Reversed string is:", rev_str)

if str == rev_str:
    print("The string is a palindrome")
else:
    print("Not a palindrome...")
```

# word frequency using strings

```python
#word frequency using strings
text = input("Enter a string: ")
words = text.split()
word_freq = {}
for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1
print("Word Frequency:", word_freq)
```

# Arrays and list

Check duplicates in list

```python
def has_duplicates(ls):
    if len(ls) != len(set(ls)): 
        return 1
    return 0 

my_ls = [1,2,3,4,4,5]
print(has_duplicates(my_ls))
```

Rotate by k

```python
def rotate(arr, k):
    n = len(arr)
    k= k%n
    return arr[k:] + arr[:k]
my_array = [1,2,3,4,5]
k = 1
print(rotate(my_array, k))
```
