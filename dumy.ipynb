```python
def merge_arr(arr1, arr2):
    merged = []
    i= 0
    j=0
    while i< len(arr1) and j< len(arr2):
        if arr1[i] <arr2[j]:
            merged.append(arr1[i])
            i+=1
        else:
            merged.append(arr2[j])
            j+=1
    while(i<len(arr1)):
        merged.append(arr1[1])
        i+=1
    while(j<len(arr2)):
        merged.append(arr2[j])
        j+=1
    return merged


ar1 = [1,2,3,6]
ar2 = [2,5,7]
print(merge_arr(ar1, ar2))
```
